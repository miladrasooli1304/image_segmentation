```python
image_org = cv2.cvtColor(image_org,cv2.COLOR_BGR2RGB)
```

```python
image_vec = image_org.reshape(-1,3)
image_vec
```

```python
image_vec.shape
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(image_vec)
```

```python
np.unique(kmeans.labels_,return_counts=True)
```

```python
labels = kmeans.labels_
labels
```

```python
imagggg= labels.reshape(image_org.shape[:2])
imagggg
```
